# 7. Классификатор ближайших соседей

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
```

```python
classifier = KNeighborsClassifier()
```

```python
classifier.fit(X_train, Y_train)
```

```python
pred = classifier.predict(X_test)
real = Y_test
```

```python
confmatr = confusion_matrix(real, pred)
```

```python
confmatr
```

Результаты уже очень хорошие, неверно предсказаны лишь 6 экзмепляров из 2031. Попробуем подобрать наилучшее число соседей. Так как нет несбалансированных классов, можно оценивать качество по accuracy.

```python
neighbors = list(range(1, 50))
```

```python
min_metric=10000
min_n=0

for neighbor in neighbors:
    print(neighbor, "neighbors fitting...")
    classifier = KNeighborsClassifier(n_neighbors=neighbor, n_jobs=-1)
    classifier.fit(X_train, Y_train)
    pred = classifier.predict(X_test)
    real = Y_test
    metric = (confusion_matrix(real, pred)[0,1] + confusion_matrix(real, pred)[1,0])/len(X_test)
    if (metric < min_metric):
        min_metric = metric
        min_n=neighbor
```

```python
min_metric
```

```python
min_n
```

Теперь посчитаем ошибку на трейне и тесте.

```python
classifier = KNeighborsClassifier(n_neighbors=1)
classifier.fit(X_train, Y_train)
test_pred = classifier.predict(X_test)
test_real = Y_test

train_pred = classifier.predict(X_train)
train_real = Y_train
```

```python
print("Test confusion matrix and error: ")
print(confusion_matrix(test_real, test_pred), "error: ", (confusion_matrix(test_real, test_pred)[0,1] + confusion_matrix(test_real, test_pred)[1,0])/len(X_test))
print("Train confusion matrix and error: ")
print(confusion_matrix(train_real, train_pred) , "error: ", (confusion_matrix(train_real, train_pred)[0,1] + confusion_matrix(train_real, train_pred)[1,0])/len(X_train))
```

Ошибка на трейне 0.0, на тесте 0.0009. И на трейне, и на тесте ошибка очень маленькая, значит можно говорить, что мы получили хорошую модель, которая не старадает переобучением!
